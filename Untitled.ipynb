```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('alzheimerpt3.txt', sep='\t', lineterminator='\n')
```

```python
df.tail(92)
```

```python
df.head(25)
```

```python
df.drop(['Notes'], axis=1)
```

```python
n = 91
 
# Removing last n rows
df_dropped = df.iloc[:-n]
 
```

```python
df_dropped.drop(['Notes'], axis=1)
```

```python
df_dropped.drop(['Notes'], axis=1)
```

```python
df_dropped['Race'].hist(bins = 25,
                            color = 'coral',
                            edgecolor = 'black',
                            figsize = (10,6))

plt.title('Ethnicity', fontsize = 16)
plt.xlabel('Race')
plt.ylabel('Number of Deaths per thousand')
plt.savefig('ethnicity2.pdf');
```

```python
df_dropped['Gender'].hist(bins = 25,
                            color = 'coral',
                            edgecolor = 'black',
                            figsize = (10,6))

plt.title('Gender', fontsize = 16)
plt.xlabel('Gender')
plt.ylabel('Number of Deaths per thousand')
plt.savefig('gen2.pdf');
```
