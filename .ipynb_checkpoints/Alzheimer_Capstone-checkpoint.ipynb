```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
alz = pd.read_csv('../Alzheimer_Capstone/Alzheimer_Data.csv')
alz.head(35)
```

```python
alz.rename(columns = {'LocationDesc':'State', 'LocationAbbr':'State_Abbr',
                              'StratificationID2':'Race'}, inplace = True)
```

```python
alz
```

```python
alz1 = alz.drop(["RowId","Response","ResponseID","Low_Confidence_Limit", "High_Confidence_Limit" ,"Stratification3",
                 "StratificationCategory3", "StratificationCategoryID3", "StratificationID3", "Report"], axis=1)
alz1
```

```python
alz_df_men = alz1[alz1['Race'] == 'MALE']     
```

```python
alz_df_men
```

```python
alz_df_women = alz1[alz1['Race'] == 'FEMALE'] 
```

```python
alz['State'].value_counts()
```

```python
alz_df_women['State'].value_counts().plot(kind = 'bar');
```

```python
alz1['Race'].hist(bins = 25,
                            color = 'coral',
                            edgecolor = 'black',
                            figsize = (10,6))

plt.title('Ethnicity', fontsize = 16)
plt.xlabel('Race')
plt.ylabel('Count')
;
```

```python
people_df_unk = alz1[~alz1['Race'].isin(['MALE', 'FEMALE', 'OVERALL'])]
```

```python
people_df_unk['Race'].hist(bins = 25,
                            color = 'coral',
                            edgecolor = 'black',
                            figsize = (10,6))

plt.title('Ethnicity', fontsize = 16)
plt.xlabel('Race')
plt.ylabel('Count')
plt.savefig('people_df_unk.pdf');
```
